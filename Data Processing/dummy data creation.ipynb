```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
X1, _ = make_blobs(n_samples=1010, centers=1, cluster_std=1.9, random_state=0)
ratings1 = np.array([random.choices(ratings, k=1, weights=[0,0,1,2,2,3,3,8,9])[0] for _ in range(1010)])
genre1 = ['Hip-Hop']*len(X1) 

first = np.concatenate([X1,np.array(ratings1).reshape(-1,1),np.array(genre1).reshape(-1,1)],axis=-1)


X2, _ = make_blobs(n_samples=930, centers=1, cluster_std=0.6, random_state=13)
ratings2 = np.array([random.choices(ratings, k=1, weights=[0,1,2,3,6,5,7,9,0])[0] for _ in range(930)])
genre2 = ['Jazz']*len(X2) 
second = np.concatenate([X2,np.array(ratings2).reshape(-1,1),np.array(genre2).reshape(-1,1)],axis=-1)


X3, _ = make_blobs(n_samples=725, centers=1, cluster_std=0.4, random_state=9)
ratings3 = np.array([random.choices(ratings, k=1, weights=[0,0,3,3,3,4,8,11,9])[0] for _ in range(725)])
genre3 = ['Pop']*len(X3) 
third = np.concatenate([X3,np.array(ratings3).reshape(-1,1),np.array(genre3).reshape(-1,1)],axis=-1)


X4, _ = make_blobs(n_samples=355, centers=1, cluster_std=0.5, random_state=7)
ratings4 = np.array([random.choices(ratings, k=1, weights=[0,1,2,9,6,3,2,1,0])[0] for _ in range(355)])
genre4 = ['Blues']*len(X4) 
fourth = np.concatenate([X4,np.array(ratings4).reshape(-1,1),np.array(genre4).reshape(-1,1)],axis=-1)


X5, _ = make_blobs(n_samples=840, centers=1, cluster_std=1.3, random_state=8)
ratings5 = np.array([random.choices(ratings, k=1, weights=[3,4,4,6,7,9,8,2,1])[0] for _ in range(840)])
genre5 = ['Classical']*len(X5)
fifth = np.concatenate([X5,np.array(ratings5).reshape(-1,1),np.array(genre5).reshape(-1,1)],axis=-1)
```

```python
import random
import time
    
def str_time_prop(start, end, time_format, prop):
    """Get a time at a proportion of a range of two formatted times.

    start and end should be strings specifying times formatted in the
    given format (strftime-style), giving an interval [start, end].
    prop specifies how a proportion of the interval to be taken after
    start.  The returned time will be in the specified format.
    """

    stime = time.mktime(time.strptime(start, time_format))
    etime = time.mktime(time.strptime(end, time_format))

    ptime = stime + prop * (etime - stime)

    return time.strftime(time_format, time.localtime(ptime))


def random_date(start, end, prop):
    return str_time_prop(start, end, '%m/%d/%Y %I:%M %p', prop)
```

```python
dataframe = pd.DataFrame(np.concatenate([first,second,third,fourth,fifth]),columns = ['enc1','enc2', 'rating', 'genre'])
timestamps = [random_date("2/1/2023 12:00 AM", "2/7/2023 12:00 AM", random.random()) for i in range(len(dataframe*2)//10)] + [random_date("2/7/2023 12:00 AM", "2/14/2023 12:00 AM", random.random()) for i in range((len(dataframe)*2)//10)] + [random_date("2/14/2023 12:00 AM", "2/21/2023 12:00 AM", random.random()) for i in range((len(dataframe)*3)//10)] + [random_date("2/21/2023 12:00 AM", "2/28/2023 12:00 AM", random.random()) for i in range((len(dataframe)*3)//10)]
timestamps += [random_date("2/25/2023 12:00 AM", "2/28/2023 12:00 AM", random.random()) for i in range(len(dataframe) - len(timestamps))]
dataframe.index = pd.to_datetime(timestamps)
dataframe = dataframe.sort_index()
dataframe[['enc1','enc2']] = dataframe[['enc1','enc2']].astype(float)
```

## For the parallel Co-ordinate chart

```python
parameters ={
    'Encoder Architecture': ['VQ-VAE','VAE'],
    'Latent Dimensions':['5K','15K','25K'],
    'Noise Schedule': random.sample(list(np.arange(1e-5,1e-3,1e-5)),k=15),
    'No. Steps': [100,300,500,1000]
}
```

```python
import itertools
```

```python
value_list = list(parameters.values())
df = pd.DataFrame(list(itertools.product(*value_list)),columns=parameters.keys())
```

```python
df['Loss'] = [0.5]*len(df)
```

```python
df.loc[df['Encoder Architecture'] == 'VQ-VAE','Loss'] = df[df['Encoder Architecture'] == 'VQ-VAE']['Loss'].apply(lambda a : a - np.random.normal(0.08,scale=0.01,size=None))
df.loc[df['Encoder Architecture'] == 'VAE','Loss'] = df[df['Encoder Architecture'] == 'VAE']['Loss'].apply(lambda a : a - np.random.normal(0.02,scale=0.03,size=None))
```

```python
df.loc[df['Latent Dimensions'] == '5K','Loss'] = df[df['Latent Dimensions'] == '5K']['Loss'].apply(lambda a : a - np.random.normal(0.03,scale=0.01,size=None))
df.loc[df['Latent Dimensions'] == '15K','Loss'] = df[df['Latent Dimensions'] == '15K']['Loss'].apply(lambda a : a - np.random.normal(0.07,scale=0.03,size=None))
df.loc[df['Latent Dimensions'] == '25K','Loss'] = df[df['Latent Dimensions'] == '25K']['Loss'].apply(lambda a : a - np.random.normal(0.02,scale=0.01,size=None))
```

```python
df.loc[df['Noise Schedule'] >= 5e-4,'Loss'] = df[df['Noise Schedule'] >= 5e-4]['Loss'].apply(lambda a : a - np.random.normal(0.0001,scale=0.001,size=None))
df.loc[df['Noise Schedule'].between(1e-4,5e-4),'Loss'] = df[df['Noise Schedule'].between(1e-4,5e-4)]['Loss'].apply(lambda a : a - np.random.normal(0.0003,scale=0.001,size=None))
df.loc[df['Noise Schedule'] <= 1e-4,'Loss'] = df[df['Noise Schedule'] <= 1e-4]['Loss'].apply(lambda a : a - np.random.normal(0.0008,scale=0.005,size=None))
```

```python
df.loc[df['No. Steps'] == 100,'Loss'] = df[df['No. Steps'] == 100]['Loss'].apply(lambda a : a - np.random.normal(0.001,scale=0.001,size=None))
df.loc[df['No. Steps'] == 300,'Loss'] = df[df['No. Steps'] == 300]['Loss'].apply(lambda a : a - np.random.normal(0.004,scale=0.002,size=None))
df.loc[df['No. Steps'] == 500,'Loss'] = df[df['No. Steps'] == 500]['Loss'].apply(lambda a : a - np.random.normal(0.006,scale=0.002,size=None))
df.loc[df['No. Steps'] == 1000,'Loss'] = df[df['No. Steps'] == 1000]['Loss'].apply(lambda a : a - np.random.normal(0.01,scale=0.0001,size=None))
```

```python
df['Inverse Loss'] = 1/df['Loss']
```

```python
df['Latent Dimensions'] = df['Latent Dimensions'].apply(lambda a : int(a[:-1])*1000)
df['Vector Quantization'] = (df['Encoder Architecture']=='VQ-VAE').astype(int)
```

```python
df.columns
```

```python
import plotly.express as px
fig = px.parallel_coordinates(df, color="Inverse Loss",
                              dimensions=['Vector Quantization','Latent Dimensions', 'Noise Schedule',
       'No. Steps', 'Inverse Loss'],
                              color_continuous_scale=px.colors.diverging.PiYG,)
fig.show()
```
